## Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.

```python
df.isnull().sum()
```

```python
# Rating, Andriod Ver, Current Ver and Type has null values 
# Below are few affect that can be caused by missing/null valuse
```

```python
Biased Analysis: 
    If data with missing values is not handled properly, it can introduce bias into the analysis and model predictions. 
    Models might learn patterns from the available data but fail to generalize well to new data.

Incomplete Information: 
    Missing values can result in incomplete information, leading to an inaccurate representation of the underlying relationships in the data. 
    This can impact the model's ability to make accurate predictions.

Reduced Sample Size: 
    Rows with missing values are often removed, reducing the effective sample size and potentially leading to loss of valuable information. 
    This can impact the statistical power of the model.

Incorrect Inferences: 
    Missing values can lead to incorrect inferences and conclusions drawn from the data. Imputing missing values inappropriately can distort
    the original distribution and lead to wrong interpretations.

Increased Error Rates: 
    Models trained on data with missing values might have higher error rates when applied to new data, as they may not generalize well to unseen 
    scenarios.

Spurious Correlations: 
    Missing values can result in spurious correlations between variables, leading to false assumptions about the relationships between features.

Imputed Information: 
    Imputing missing values can introduce artificial information that may not accurately represent the true underlying values. 
    This can lead to overfitting if not handled carefully.
```

## Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
df['Size'].unique()
```

```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].str.replace('+','')
df['Size']=df['Size'].str.replace(',','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df['Size']=df['Size'].astype(float)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.ylim(0, 6)
#plt.xlim(1, 6)
sns.scatterplot(x=df['Size'],y=df['Rating'],color='blue', marker='o')
```

## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df['Type'].value_counts()
```

```python
df[df['Type'] == '0']
```

```python
df.drop(df.index[10472],inplace=True)
```

```python
df['Type'].value_counts()
```

```python
df['Price'].unique()
```

```python
# Remove any rows with missing values in 'Price' column
df = df.dropna(subset=['Price'])

# Convert 'Price' column to numeric (remove the "$" sign and convert to float)
df['Price'] = df['Price'].apply(lambda x: float(x[1:] if x[0]=='$' else x))

# Group by 'Type' and calculate the average price for each app type
avg_prices_by_type = df.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
avg_prices_by_type.plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')

# Rotate x-labels for better readability
plt.xticks(rotation=0)

# Show the plot
plt.show()
```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'].unique()
```

```python
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
```

```python
# Group by app and sum the installs for each app
app_installs = df.groupby('App')['Installs'].sum()

# Sort the app_installs Series in descending order
sorted_app_installs = app_installs.sort_values(ascending=False)

# Create a frequency table for apps with the most installs
frequency_table = sorted_app_installs.head(10)

print(frequency_table)
```

## Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
df.head()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['Day']=df['Last Updated'].dt.day
df['Month']=df['Last Updated'].dt.month
df['Year']=df['Last Updated'].dt.year
```

```python
df.drop('Last Updated', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Reviews']=df['Reviews'].astype(int)
```

```python
# To analyse further we need to extract numeric and categorical data separately 
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
```

```python
print(categorical_features)
print(numeric_features)
```

```python
## Visualization Diagrams

plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```

## Observations:
 1. From numerical feature graph we can say that graph for rating and year is left skewed.
