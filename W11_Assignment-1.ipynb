## Q1. What is the Probability density function?


```python
In probability theory and statistics, the Probability Density Function (PDF) is a fundamental concept used to describe the 
likelihood of a continuous random variable taking on a specific value within a given range. The PDF provides a way to model and 
understand the distribution of a continuous random variable.

For a continuous random variable, the PDF is denoted as f(x), where "x" represents the variable's value. 
The PDF does not directly give the probability of a single point but rather describes the probability density or concentration of the 
variable around a particular value. The probability of the random variable falling within a specific interval can be obtained by integrating 
the PDF over that interval.

The properties of a valid PDF are as follows:

    Non-negativity: The PDF is always non-negative, i.e., f(x) ≥ 0 for all x.

    Total area under the curve: The total area under the PDF curve over the entire range of the random variable is equal to 1. 
    This indicates that the variable must take on some value within its possible range.

    Mathematically, the probability of a continuous random variable X falling within an interval [a, b] can be calculated as the integral of
    the PDF over that interval:

P(a ≤ X ≤ b) = ∫(a to b) f(x) dx

The concept of the PDF is essential in various statistical analyses, such as finding expected values, computing percentiles, and understanding 
the behavior of continuous random variables in general. It is a cornerstone in probability and statistics and plays a vital role in many 
real-world applications across various fields.
```

```python
Let's consider an example of a continuous random variable following a standard normal distribution. The PDF for a standard normal 
distribution is given by the following formula:

f(x) = (1 / √(2π)) * e^((-x^2) / 2)

Here, "e" represents Euler's number, and π is the mathematical constant pi (approximately 3.14159). The standard normal distribution has
a mean (μ) of 0 and a standard deviation (σ) of 1.

Now, let's plot the PDF of the standard normal distribution and calculate the probability of the random variable falling within a specific 
interval.

Example: Finding the probability of a standard normal variable falling within [-1, 1]

Step 1: Calculate the PDF for the standard normal distribution using the formula:

f(x) = (1 / √(2π)) * e^((-x^2) / 2)

Step 2: Calculate the probability of the random variable falling within the interval [-1, 1]:

P(-1 ≤ X ≤ 1) = ∫(-1 to 1) f(x) dx

Since the standard normal distribution is symmetric, the probability of falling within [-1, 1] is the same as falling within [1, -1].
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the PDF for the standard normal distribution
def standard_normal_pdf(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Create an array of x-values from -3 to 3 (for visualization)
x = np.linspace(-3, 3, 1000)

# Calculate the PDF values for the x-values
pdf_values = standard_normal_pdf(x)

# Plot the PDF curve
plt.plot(x, pdf_values, label='PDF: Standard Normal Distribution')
plt.fill_between(x, pdf_values, 0, where=(-1 <= x) & (x <= 1), color='gray', alpha=0.5, label='Probability Area')

# Highlight the area representing P(-1 ≤ X ≤ 1)
plt.text(0, 0.1, 'P(-1 ≤ X ≤ 1)', ha='center', fontsize=12)

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Probability Density Function of Standard Normal Distribution')
plt.grid(True)
plt.show()
```

## Q2. What are the types of Probability distribution?


```python
There are several types of probability distributions, each serving different purposes and modeling different types of 
random variables. 
Probability distributions can be broadly categorized into two main types: discrete probability distributions and continuous probability 
distributions.

Discrete Probability Distributions:
    
    Discrete probability distributions are used to model random variables that can take on a countable set of distinct values. 
    The probability mass function (PMF) is used to describe the probabilities of each possible value.

    Examples of discrete probability distributions include:

        Bernoulli Distribution: Models a binary random variable with two possible outcomes (e.g., success/failure, heads/tails).
        Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
        Poisson Distribution: Models the number of events occurring in a fixed interval of time or space.
        Geometric Distribution: Models the number of trials needed until the first success in a sequence of Bernoulli trials.
        Hypergeometric Distribution: Models the probability of drawing a certain number of successes in a sample without replacement.
        
Continuous Probability Distributions:
    
    Continuous probability distributions are used to model random variables that can take on any value within a continuous range. 
    The probability density function (PDF) is used to describe the likelihood of the random variable taking on a specific value.

    Examples of continuous probability distributions include:

        Normal Distribution: A symmetric bell-shaped distribution, often used to model real-world measurements that are influenced by many 
        small random factors.
        Uniform Distribution: Models a random variable that is equally likely to take on any value within a specified interval.
        Exponential Distribution: Models the time between events in a Poisson process, where events occur continuously and independently at 
        a constant rate.
        Beta Distribution: A family of distributions that is useful for modeling proportions or probabilities.
        Gamma Distribution: Generalizes the exponential distribution and is used in various applications, including queuing theory and 
        reliability analysis.

        
It is important to note that there are many other probability distributions beyond these examples, each suited to different scenarios and 
applications. The choice of which distribution to use depends on the nature of the data and the specific problem being addressed in a 
particular context. Understanding the characteristics and properties of these distributions is essential in probability theory and statistics 
for data analysis, modeling, and inference.
```

## Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.


```python
To calculate the probability density function (PDF) of a normal distribution at a given point, you can use the formula 
for the PDF of the normal distribution:

PDF(x; μ, σ) = (1 / (σ * √(2π))) * exp(-(x - μ)^2 / (2 * σ^2))

where:

x is the value at which you want to evaluate the PDF.
μ is the mean of the normal distribution.
σ is the standard deviation of the normal distribution.
π is the mathematical constant pi (approximately 3.14159).
exp() is the exponential function.
```

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution
    at the given point x.

    Args:
        x (float): The value at which to evaluate the PDF.
        mean (float): The mean (μ) of the normal distribution.
        std_dev (float): The standard deviation (σ) of the normal distribution.

    Returns:
        float: The probability density function (PDF) value at the given point x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value
```

```python
mean = 0.0
std_dev = 1.0
x_value = 1.5

pdf_at_x = normal_pdf(x_value, mean, std_dev)
print(f"PDF at x = {x_value}: {pdf_at_x}")
```

## Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.


```python
The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of
independent Bernoulli trials. The Binomial distribution has several key properties:

Fixed Number of Trials: 
    The Binomial distribution requires a fixed number of trials denoted by "n." Each trial is independent of others, and there are only two
    possible outcomes (success or failure) for each trial.

Constant Probability of Success: 
    The probability of success in each trial is denoted by "p," and it remains constant throughout all the trials. The probability of failure
    is then given by (1 - p).

Discrete Nature: 
    The Binomial distribution deals with discrete random variables since the number of successes is an integer between 0 and "n."

Notation: 
    The Binomial distribution is denoted as B(x; n, p), where "x" represents the number of successes, "n" is the total number of trials, and 
    "p" is the probability of success in each trial.

Two examples of events where the Binomial distribution can be applied are:

Coin Tosses:
    Suppose we have a fair coin, and we want to find the probability of getting exactly "k" heads in "n" tosses. Each toss has two possible
    outcomes (heads or tails) with a fixed probability of success (0.5 for a fair coin). This situation can be modeled using a Binomial
    distribution, where "n" is the number of coin tosses, "k" is the number of heads we want to observe, and "p" is the probability of getting 
    a head in a single toss.

Quality Control:
    In a manufacturing process, products may be classified as either defective or non-defective. Let's say we have a batch of "n" products, 
    and each product has a probability "p" of being defective (failure) due to a manufacturing flaw. We are interested in finding the 
    probability of having exactly "k" defective products in the batch. This scenario can be modeled using a Binomial distribution, where 
    "n" is the batch size, "k" is the number of defective products we want to observe, and "p" is the probability of a single product being 
    defective.

In both examples, we have a fixed number of trials (coin tosses or product inspections), and each trial has two possible outcomes (success or 
failure) with a constant probability of success (0.5 for coin tosses or "p" for product inspections). The number of successful outcomes 
(heads or defective products) follows a Binomial distribution.
```

## Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

def generate_binomial_sample(size, probability):
    return np.random.binomial(n=1, p=probability, size=size)

# Parameters for the binomial distribution
sample_size = 1000
probability_of_success = 0.4

# Generate the random sample
sample = generate_binomial_sample(sample_size, probability_of_success)

# Plot the histogram
plt.hist(sample, bins=2, alpha=0.6, rwidth=0.8, color='blue')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (n={sample_size}, p={probability_of_success})')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
In this code, we define a function generate_binomial_sample that uses numpy's random.
binomial function to generate a random sample from the binomial distribution. The size parameter indicates the number of trials in the 
sample, and probability represents the probability of success in each trial.

We then specify the desired sample size (sample_size) and the probability of success (probability_of_success) for the binomial distribution. 
We call the generate_binomial_sample function with these parameters to create the random sample.

Finally, we plot the histogram using matplotlib.pyplot.hist. The histogram will display the frequency of successes (1s) and failures (0s) in 
the generated sample. The xticks and labels are added to make the plot more informative.
```

## Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.


```python
To calculate the Cumulative Distribution Function (CDF) of a Poisson distribution at a given point, you can use the 
formula for the CDF of the Poisson distribution:

CDF(k; λ) = ∑(i=0 to k) ((λ^i) * exp(-λ)) / i!

where:

k is the point at which you want to evaluate the CDF.
λ (lambda) is the mean of the Poisson distribution.
exp() is the exponential function.
```

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution
    at the given point k.

    Args:
        k (int): The value at which to evaluate the CDF.
        mean (float): The mean (λ) of the Poisson distribution.

    Returns:
        float: The cumulative distribution function (CDF) value at the given point k.
    """
    cdf_value = 0.0
    for i in range(k + 1):
        cdf_value += (mean**i * math.exp(-mean)) / math.factorial(i)
    return cdf_value

mean = 2.5
point = 3

cdf_at_point = poisson_cdf(point, mean)
print(f"CDF at k = {point}: {cdf_at_point}")
```

## Q7. How Binomial distribution different from Poisson distribution?


```python
The Binomial distribution and the Poisson distribution are both probability distributions used to model discrete random 
variables. However, they have distinct characteristics and are suitable for different types of scenarios. 
Here are the key differences between the two distributions:

Nature of Events:

    Binomial Distribution: 
        The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, 
        where each trial has two possible outcomes (success or failure) with a constant probability of success (p). The trials are mutually 
        exclusive and independent.
    
    Poisson Distribution: 
        The Poisson distribution models the number of events that occur in a fixed interval of time or space. It is used 
        to describe rare events that happen randomly and independently at a constant average rate (λ).

Number of Trials/Events:

    Binomial Distribution: 
        The Binomial distribution requires a fixed number of trials (n) and deals with a discrete number of successful outcomes (k) within 
        these trials.
        
    Poisson Distribution: 
        The Poisson distribution deals with a continuous range of events, and there is no fixed number of trials. It represents the number 
        of occurrences of an event in a continuous time or space interval.
    
Parameters:

    Binomial Distribution: 
        The Binomial distribution is characterized by two parameters: the number of trials (n) and the probability of success in each trial (p).
        
    Poisson Distribution: 
        The Poisson distribution is characterized by one parameter: the average rate (λ) of events occurring in the fixed interval.

Limiting Case:

    Binomial Distribution: 
        The Binomial distribution approaches the Poisson distribution as the number of trials (n) becomes large, and the probability of 
        success (p) becomes small, while keeping the product np constant.
        
    Poisson Distribution: 
        The Poisson distribution can be seen as a limiting case of the Binomial distribution when the number of trials becomes very large, 
        and the probability of success in each trial becomes very small.

Shape of Distribution:

    Binomial Distribution: The Binomial distribution is typically bell-shaped and symmetric when the number of trials is reasonably large.
    
    Poisson Distribution: The Poisson distribution is skewed to the right and often represents a small number of rare events.

In summary, the key difference between the Binomial and Poisson distributions lies in the nature of the events they model. The Binomial 
distribution is used when there are a fixed number of independent trials with two possible outcomes, while the Poisson distribution is 
employed when dealing with rare and random events occurring continuously in time or space.
```

## Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.


```python
import numpy as np

def generate_poisson_sample(size, mean):
    """
    Generate a random sample from a Poisson distribution.

    Args:
        size (int): The size of the sample.
        mean (float): The mean of the Poisson distribution.

    Returns:
        numpy.ndarray: The generated random sample.
    """
    return np.random.poisson(mean, size=size)

# Parameters for the Poisson distribution
sample_size = 1000
mean_poisson = 5

# Generate the random sample from the Poisson distribution
sample_poisson = generate_poisson_sample(sample_size, mean_poisson)

# Calculate the sample mean and variance
sample_mean = np.mean(sample_poisson)
sample_variance = np.var(sample_poisson)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

```python
In this code, we define a function generate_poisson_sample that uses numpy's random.
poisson function to generate a random sample from the Poisson distribution. The size parameter indicates the number of samples in 
the generated sample, and mean represents the mean of the Poisson distribution.

We then specify the desired sample size (sample_size) and the mean of the Poisson distribution (mean_poisson). 
We call the generate_poisson_sample function with these parameters to create the random sample.

Finally, we calculate the sample mean and variance using numpy's mean and var functions, respectively, and print the results. 
The sample mean should be close to the specified mean of 5, and the sample variance should be close to the square of the mean 
(i.e., 5^2 = 25) as per the properties of the Poisson distribution.
```

## Q9. How mean and variance are related in Binomial distribution and Poisson distribution?


```python
In both the Binomial distribution and the Poisson distribution, the mean and variance are related, but the relationship
differs between the two distributions.

Binomial Distribution:
For a Binomial distribution with parameters n (number of trials) and p (probability of success in each trial), the mean (μ) and 
variance (σ^2) are related as follows:

Mean (μ) = n * p
Variance (σ^2) = n * p * (1 - p)

Here, the mean represents the expected number of successes in the n trials, and the variance measures the spread or dispersion of the 
distribution. The variance is highest when p = 0.5 (the distribution is the most spread out), and it decreases as p moves towards 0 or 1 
(the distribution becomes more concentrated around the mean).

Poisson Distribution:
For a Poisson distribution with a parameter λ (the average rate of events occurring in a fixed interval), the mean (μ) and variance (σ^2) 
are related as follows:

Mean (μ) = λ
Variance (σ^2) = λ

In the case of the Poisson distribution, the mean and variance are both equal to the parameter λ. This property makes the Poisson distribution
unique, as the variance does not depend on any other parameter. The Poisson distribution is often used to model rare events, and its variance 
is equal to its mean, which simplifies some calculations and makes it easier to work with in certain scenarios.

In summary:

In the Binomial distribution, the variance is influenced by both the number of trials (n) and the probability of success (p).
In the Poisson distribution, the variance is solely determined by the average rate (λ) of events.
```

## Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

```python

In a normal distribution, the least frequent data points appear in the tails of the distribution, which are located farthest from the mean.

A normal distribution is a symmetric, bell-shaped probability distribution that is fully characterized by its mean (μ) and standard deviation 
(σ). The mean is the central value of the distribution and represents the peak or highest point of the bell-shaped curve.

As you move away from the mean towards the tails of the distribution, the frequency of data points decreases rapidly. The tails represent 
the extreme values, both on the left (negative side) and the right (positive side) of the mean. These extreme values are less likely to occur 
compared to values closer to the mean.

The reason for this phenomenon lies in the nature of the normal distribution. It is defined such that the probability density decreases as 
we move away from the mean. The tails extend to infinity, and the curve approaches but never touches the horizontal axis.

In summary, the least frequent data points in a normal distribution are found in the tails, which are farthest away from the mean. 
The values closer to the mean are more common, resulting in the characteristic bell-shaped curve of the normal distribution.
```
