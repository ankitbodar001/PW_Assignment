## Q8. You are a data scientist working for an airline company, and you have been asked to analyze the Flight Price dataset to identify the factors that affect flight prices. What features would you analyze to identify these factors, and how would you present your findings to the management team?

```python
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df['Arrival_hours']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Arrival_hours']=df['Arrival_hours'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df['Dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dept_min']=df['Dep_Time'].str.split(':').str[1]
df['Dept_hour']=df['Dept_hour'].astype(int)
df['Dept_min']=df['Dept_min'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Duration_hour']=df['Duration_hour'].str.split('m').str[0]
df['Duration_hour']=df['Duration_hour'].astype(int)
```

```python
df['Total_Stops'].unique()
```

```python
import numpy as np
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
print(df['Airline'].unique())
print(df['Source'].unique())
print(df['Destination'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df,encoded_df],axis=1)
```

```python
df.head()
```

```python
df.drop(['Airline','Source','Destination','Additional_Info'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation = 90)
sns.set(rc = {'figure.figsize':(16,8)})
sns.heatmap(df[['Price','Airline_Air Asia','Airline_Jet Airways','Airline_Air India','Airline_GoAir','Airline_IndiGo','Airline_Jet Airways Business',
                'Airline_Multiple carriers','Airline_Multiple carriers Premium economy','Airline_SpiceJet','Airline_Trujet','Airline_Vistara',
                'Airline_Vistara Premium economy']].corr(), 
            annot = True, fmt='.2g',cmap= 'coolwarm')
```

## from above we can say that Airline brand does not have any specific effect on the flight prices, there is positive correlation
## between Jet Airways and price but it does not have significant effect on Prices
